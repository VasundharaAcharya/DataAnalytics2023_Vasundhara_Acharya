```r
# Let us experiment with different independant variables


mm1<- lm(Homeowners~area)
summary(mm1)
```

```r
mm1<- lm(Homeowners~area+Immigrant)
summary(mm1)
```

```r
plot(area, Homeowners)
abline(mm1)
abline(mm1,col=3,lwd=3) # line color = green, line width = 3
```

```r
##########The independant variable here is the Income
mm2<- lm(Homeowners~Income)
summary(mm2)
```

```r
plot(Income, Homeowners)
abline(mm2)
abline(mm2,col=7,lwd=3) # line color = green, line width = 3
```

```r
##########PLOTTING THE RESIDUAL###########
model_residuals = mm2$residuals

# Plot the result
hist(model_residuals)
```

```r
#######If there is normality it must be following straight line
#Here we see that it doesnt follow a straight line, hence its not normal distribution
# Plot the residuals
qqnorm(model_residuals)
# Plot the Q-Q line
qqline(model_residuals)
```

```r
###########Multiple independant variables#######
model = lm(formula = Homeowners ~ Income+area+Population)
```

```r
model_residuals = model$residuals

# Plot the result
hist(model_residuals)
```

```r
#######If there is normality it must be following straight line
#Here we see that it almost follows a straight line
# Plot the residuals
qqnorm(model_residuals)
# Plot the Q-Q line
qqline(model_residuals)
#####From the resultant we see that almost all points follow a straight line. Hence it can be normal distribution
```
